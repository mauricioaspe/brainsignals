```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
ID = 'SERT' + '1984'
structure = 'vHip' + '_epochs'
```

```python
npys_dir  = '/home/maspe/filer/SERT/' + ID + '/npys/'

file_name = npys_dir + structure + '.npy'
print('Loading {}...'.format(file_name))
matrix = np.load(file_name, allow_pickle=True)
print('Matrix shape = {}'.format(matrix.shape))
```

########

```python
channel = 6
epoch = 8
print('Selected channel {}, epoch {}, in {} from {}'. format(channel, epoch, structure[:-7], ID))
```

```python
plt.plot(matrix[channel, :, epoch-1])
plt.title('Bad: channel {}, epoch {}, in {} from {}'.format(channel, epoch, structure[:-7], ID));
```

```python
print('Deleting epoch {} in channel {}...'.format(epoch, channel))
matrix[channel, :, epoch-1] = matrix[channel, :, epoch-2]
print('New matrix shape = {}'.format(matrix.shape))

plt.plot(matrix[channel, :, epoch-1])
plt.title('Fixed: channel {}, epoch {}, in {} from {}'.format(channel, epoch, structure[:-7], ID));
```

###########

```python
all_channels = True
epoch = 8
```

```python
if all_channels:
    n_channels = matrix.shape[0]
    plt.figure(figsize=(10,20))
    plt.subplots_adjust(hspace=0.6)

    for chan in range(n_channels):
        plt.subplot(n_channels, 1, chan+1)
        plt.title('Bad: channel {}, epoch {}, in {} from {}'.format(chan, epoch, structure[:-7], ID))
        plt.plot(matrix[chan, :, epoch-1])
        #plt.title('Fixed: channel {}, epoch {}, in {} from {}'.format(channel, epoch, structure[:-7], ID));
    
all_channels = False
```

```python
all_channels = True
epoch = 8
```

```python
if all_channels:
    print('Deleting all channels in epoch {}...'.format(epoch))
    matrix[:, :, epoch-1] = matrix[:, :, epoch-2]
    print('New matrix shape = {}'.format(matrix.shape))

    n_channels = matrix.shape[0]

    plt.figure(figsize=(10,20))
    plt.subplots_adjust(hspace=0.6)
    for chan in range(n_channels):
        plt.subplot(n_channels, 1, chan+1)
        plt.title('Good: channel {}, epoch {}, in {} from {}'.format(chan, epoch, structure[:-7], ID))
        plt.plot(matrix[chan, :, epoch-1])
        #plt.title('Fixed: channel {}, epoch {}, in {} from {}'.format(channel, epoch, structure[:-7], ID));
    
all_channels = False
```
