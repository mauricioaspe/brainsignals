#### Create butterworth filter and plot its frequency response

```python
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = signal.butter(order, [low, high], btype='band')
    return b, a

#N       = 2
fs      = 1000.0
lowcut  = 5.0
highcut = 12.0

N = 4
b, a = butter_bandpass(lowcut, highcut, fs, order=N)
theta_WT = signal.filtfilt(b=b, a=a, x=mPFC_WT,
                           axis=-1, padtype='odd', padlen=None, method='pad', irlen=None)

theta_KO = signal.filtfilt(b=b, a=a, x=mPFC_KO,
                           axis=-1, padtype='odd', padlen=None, method='pad', irlen=None)
```

#### Apply Hilbert transform to form the analytic signal

```python
z_WT = signal.hilbert(theta_WT)
z_KO = signal.hilbert(theta_KO)
```
