#### Select data

```python
# Selecting mPFC
data = mean_mPFC
sampleRate = 30000
```

#### Band-pass signal and plot power spectral density

```python
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = signal.butter(order, [low, high], btype='band')
    return b, a

N       = 2
fs      = sampleRate
lowcut  = 4.0
highcut = 12.0

b, a = butter_bandpass(lowcut, highcut, fs, order=N)
```

```python
theta_band = signal.filtfilt(b=b, a=a, x=data - np.mean(data),
                           axis=-1, padtype='odd', padlen=None, method='pad', irlen=None)
```

#### Filtering plot (CHANGE FOR PSD!!)

```python
fig = plt.figure()
plt.subplot(2,1,1)
plt.plot(data[10000:40000])
plt.subplot(2,1,2)
plt.plot(theta_band[10000:40000])

# fig.savefig("/home/maspe/filer/testFiles/figs/theta.png", dpi=150, format='png')
```

#### Apply Hilbert transform to form the analytic signal

```python
z = signal.hilbert(theta_band)
```

```python
z.imag
```

```python
phases = []
amplitudes = []
total_units = len(all_spikes)

for unit in range(total_units):
    phases.append(z.imag[np.asarray(all_spikes[unit])])
    amplitudes.append(z.real[np.asarray(all_spikes[unit])])
```

```python
flat_amplitudes = [item for sublist in amplitudes for item in sublist]
flat_phases = [item for sublist in phases for item in sublist]
```

```python
plt.polar(flat_amplitudes, flat_phases, 'go')
```
