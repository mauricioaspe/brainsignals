```python
from scipy.signal import bessel, butter, lfilter
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz
```

```python
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

fs = 1000.0

# Delta band
delta_lowcut = 0.5
delta_highcut = 3.0

# Theta band
theta_lowcut = 3.5
theta_highcut = 12.0 #7.5

# Alpha band
alpha_lowcut = 8.0
alpha_highcut = 12.0

# Beta band
beta_lowcut = 13.0
beta_highcut = 20.0

# Low gamma band
lgamma_lowcut = 21.0
lgamma_highcut = 50.0

# High gamma band
hgamma_lowcut = 51.0
hgamma_highcut = 80.0

# Gamma low_pass
lpgamma_highcut = 80.0

### Plot the frequency response for a few different orders.
# Delta
plt.figure(figsize=(10, 20))
plt.subplot(6,1,1)
for order in [1, 2, 3, 4]:
    b, a = butter_bandpass(delta_lowcut, delta_highcut, fs, order=order)
    w, h = freqz(b, a, worN=512)
    plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)], '--', label='sqrt(0.5)')

plt.axvline(x=delta_lowcut, color='black')
plt.axvline(x=delta_highcut, color='black')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.xlim([0,10])
plt.grid(True)
plt.legend(loc='best')
plt.title('Delta')
plt.show()

# Theta
plt.figure(figsize=(10, 20))
plt.subplot(6,1,2)
for order in [2, 3, 4, 5]:
    b, a = butter_bandpass(theta_lowcut, theta_highcut, fs, order=order)
    w, h = freqz(b, a, worN=512)
    plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)],
         '--', label='sqrt(0.5)')

plt.axvline(x=theta_lowcut, color='black')
plt.axvline(x=theta_highcut, color='black')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.xlim([0,25])
plt.grid(True)
plt.legend(loc='best')
plt.title('Theta')
plt.show()

# Alpha
plt.figure(figsize=(10, 20))
plt.subplot(6,1,3)
for order in [2, 3, 4, 5]:
    b, a = butter_bandpass(alpha_lowcut, alpha_highcut, fs, order=order)
    w, h = freqz(b, a, worN=512)
    plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)],
         '--', label='sqrt(0.5)')

plt.axvline(x=alpha_lowcut, color='black')
plt.axvline(x=alpha_highcut, color='black')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.xlim([0,25])
plt.grid(True)
plt.legend(loc='best')
plt.title('Alpha')
plt.show()

# Beta
plt.figure(figsize=(10, 20))
plt.subplot(6,1,4)
for order in [4, 5, 6, 7]:
    b, a = butter_bandpass(beta_lowcut, beta_highcut, fs, order=order)
    w, h = freqz(b, a, worN=512)
    plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)],
         '--', label='sqrt(0.5)')

plt.axvline(x=beta_lowcut, color='black')
plt.axvline(x=beta_highcut, color='black')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.xlim([0,25])
plt.grid(True)
plt.legend(loc='best')
plt.title('Beta')
plt.show()

# Low gamma
plt.figure(figsize=(10, 20))
plt.subplot(6,1,5)
for order in [4, 5, 6, 7]:
    b, a = butter_bandpass(lgamma_lowcut, lgamma_highcut, fs, order=order)
    w, h = freqz(b, a, worN=512)
    plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)],
         '--', label='sqrt(0.5)')

plt.axvline(x=lgamma_lowcut, color='black')
plt.axvline(x=lgamma_highcut, color='black')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.xlim([15,60])
plt.grid(True)
plt.legend(loc='best')
plt.title('Low gamma')
plt.show()

# High gamma
plt.figure(figsize=(10, 20))
plt.subplot(6,1,6)
for order in [4, 5, 6, 7]:
    b, a = butter_bandpass(hgamma_lowcut, hgamma_highcut, fs, order=order)
    w, h = freqz(b, a, worN=512)
    plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)],
         '--', label='sqrt(0.5)')

plt.axvline(x=hgamma_lowcut, color='black')
plt.axvline(x=hgamma_highcut, color='black')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.xlim([40,100])
plt.grid(True)
plt.legend(loc='best')
plt.title('High gamma')
plt.show()


# Gamma low-pass

def butter_bandpass(highcut, fs, order=5):
    nyq = 0.5 * fs
    high = highcut / nyq
    b, a = butter(order, high, btype='lowpass')
    return b, a

fs = 1000.0


plt.figure(figsize=(10, 20))
plt.subplot(6,1,6)
for order in [15, 20, 23]:
    b, a = butter_bandpass(lpgamma_highcut, fs, order=order)
    w, h = freqz(b, a, worN=512)
    plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)],
         '--', label='sqrt(0.5)')

plt.axvline(x=hgamma_lowcut, color='black')
plt.axvline(x=hgamma_highcut, color='black')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.xlim([40,100])
plt.grid(True)
plt.legend(loc='best')
plt.title('High gamma')
plt.show()
```
