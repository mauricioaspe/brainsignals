```python
import numpy as np
import matplotlib.pyplot as plt
import Wavelets
```

```python
dt=0.01
time=np.arange(0,50,dt)

fbasal1=30
fbasal2=15
amp1=1
amp2=2
bias=55
ampruido=0.2
chrpini=15
chrpstp=30

x=bias + amp1*np.sin(2*np.pi*fbasal1*time) + amp2*np.sin(2*np.pi*fbasal2*time)
x+=np.random.normal(size=len(x))*ampruido

x2=1.2*np.cos(2*np.pi*0.8*(time-chrpini)**2) * (time-chrpini)*np.exp(-(time-chrpini)/10)
x2[(time<chrpini)+(time>chrpstp)]=0
x+=x2

plt.figure(figsize=(12,4)) 
plt.clf()
plt.plot(time,x,alpha=1)
plt.xlabel('Time (s)')
```

```python
plt.specgram(x-np.mean(x),Fs=1/dt,NFFT=128,noverlap=64);
```

```python
desiredFreq=np.arange(1,40,0.2)
desiredPeriods=1/(desiredFreq*dt)
scales=desiredPeriods/Wavelets.Morlet.fourierwl

wavel1=Wavelets.Morlet(x,scales=scales)

pwr1=wavel1.getnormpower()

fmin=min(desiredFreq)
fmax=max(desiredFreq)

plt.figure(1,figsize=(10,4))
plt.clf()

ax1=plt.subplot2grid((1,5),(0,0),colspan=4)
#
plt.imshow(pwr1,cmap='RdBu',vmax=np.max(pwr1),vmin=-np.max(pwr1),
           extent=(min(time),max(time),fmin,fmax),origin='lower', 
           interpolation='none',aspect='auto')
#ax1.set_yscale('log')
ax1.set_ylabel('Frequency (Hz)')

ax2=plt.subplot2grid((1,5),(0,4))
ax2.plot(np.sum(pwr1,-1),desiredFreq)
```

```python
dt=0.02
time=np.arange(0,20,dt)
x1=np.zeros_like(time)
x1[0:500]=5 + 3*np.sin(2*np.pi*5*time[0:500])
x1[500:]=5 + 2.2*np.sin(2*np.pi*2*time[500:])
#x1+=np.random.normal(size=len(x1))*0.4

x2=-2 + 1.5*np.sin(2*np.pi*5*time) + 1.1*np.sin(2*np.pi*2*time)
#x2+=np.random.normal(size=len(x1))*0.4

plt.clf()
plt.plot(time,x1,alpha=1)
plt.plot(time,x2,alpha=1)
plt.xlabel('Time (s)')
```

```python
fftx1=np.fft.fft(x1-np.mean(x1))
fftx2=np.fft.fft(x2-np.mean(x2))
ffreq=np.fft.fftfreq(len(x1),dt)

limplot=range(200)

plt.clf()
plt.plot(ffreq[limplot],np.abs(fftx1[limplot])**2)
plt.plot(ffreq[limplot],np.abs(fftx2[limplot])**2)
plt.xlim((0,10))
#plt.yscale('log');plt.ylim((0.1,0.5e8))

plt.xlabel('Freq (Hz)')
```

```python
plt.figure(figsize=(10,4))
plt.subplot(121)
spectrum,freqs,t,_=plt.specgram(x1-np.mean(x1),Fs=50)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')

plt.subplot(122)
spectrum,freqs,t,_=plt.specgram(x2-np.mean(x2),Fs=50)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
```

```python
desiredFreq=np.arange(1,15,0.2)
desiredPeriods=1/(desiredFreq*dt)
scales=desiredPeriods/Wavelets.Morlet.fourierwl

wavel1=Wavelets.Morlet(x1,scales=scales)
pwr1=wavel1.getnormpower()

wavel2=Wavelets.Morlet(x2,scales=scales)
pwr2=wavel2.getnormpower()
waveletT=wavel1.getdata()

fmin=min(desiredFreq)
fmax=max(desiredFreq)

plt.figure(figsize=(10,4))
ax1=plt.subplot(121)

plt.imshow(pwr1,cmap='jet',
           extent=(min(time),max(time),fmin,fmax),origin='lower', 
           interpolation='none',aspect='auto')
#ax1.set_yscale('log')
ax1.set_ylabel('Frequency (Hz)')

ax2=plt.subplot(122)

plt.imshow(pwr2,cmap='jet',
           extent=(min(time),max(time),fmin,fmax),origin='lower', 
           interpolation='none',aspect='auto')
#ax1.set_yscale('log')
ax2.set_ylabel('Frequency (Hz)')
```
